```python
#import Libraries 
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python

stop_words = list(STOP_WORDS)
```

```python
#Load NLP Model
nlp = spacy.load('en_core_web_sm')
```

```python
text = "A lion was once sleeping in the jungle when a mouse started running up and down his body just for fun. This disturbed the lion’s sleep, and he woke up quite angry. He was about to eat the mouse when the mouse desperately requested the lion to set him free. I promise you, I will be of great help to you someday if you save me. The lion laughed at the mouse’s confidence and let him go. One day, a few hunters came into the forest and took the lion with them. They tied him up against a tree. The lion was struggling to get out and started to whimper. Soon, the mouse walked past and noticed the lion in trouble. Quickly, he ran and gnawed on the ropes to set the lion free. Both of them sped off into the jungle."
```

```python
#For Tokenization 

doc = nlp(text)
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
#Stoping Stop Words And Punctuation 
#Text Cleaning
```

```python
punctuation = punctuation + '\n'
punctuation
```

```python
word_freqencies = {}
for word in doc:
  if word.text.lower() not in stop_words:
    if word.text.lower() not in punctuation:
      if word.text not in word_freqencies.keys():
        word_freqencies[word.text] = 1
      else:
        word_freqencies[word.text] += 1
```

```python
word_freqencies
```

```python
# for we want to get normalize frequency so we calculate first maximum frequency and after devide them
max_frequency = max(word_freqencies.values())
```

```python
max_frequency
```

```python
#Devide all of them by max frequency
for word in word_freqencies.keys():
  word_freqencies[word] = word_freqencies[word]/max_frequency
```

```python
word_freqencies
```

```python
#Sentance Tokenization 
sentance_token = [sent for sent in doc.sents]
```

```python
sentance_token
```

```python
#Calculate Sentance Score 
```

```python
sentance_scores = {}
for sent in sentance_token:
  for word in sent:
    if word.text.lower() in word_freqencies.keys():
      if sent not in sentance_scores.keys():
        sentance_scores[sent] = word_freqencies[word.text.lower()]
      else:
        sentance_scores[sent] += word_freqencies[word.text.lower()]
```

```python
sentance_scores
```

```python
#importing 
from heapq import nlargest
```

```python
#length 
select_length = int(len(sentance_token)*0.3)
select_length
```

```python
summary = nlargest(select_length, sentance_scores,key = sentance_scores.get)
```

```python
summary
```

```python
final_summary = [word.text for word in summary]
```

```python
summary = ' '.join(final_summary)
```

```python
print(summary)
```
